```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt

data = pd.read_csv('ahorros.csv')
data.head(11)
```

```python
# Separar variables predictivas (X) y variable objetivo (y)
X = data[['Edad', 'Ingresos', 'Gastos']]
y = data['Ahorros']
```

```python
# Escalado de los valores predictivos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Separar en conjunto de entrenamiento y prueba (80% entrenamiento, 20% prueba)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Ajustar el modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(X_train, y_train)
```

```python
#  Validar con R² en el conjunto de prueba
y_pred = modelo.predict(X_test)
r2_antes = r2_score(y_test, y_pred)
print(f'R² en conjunto de prueba: {r2_antes}')
```

```python
cv_scores = cross_val_score(modelo, X_scaled, y, cv=5, scoring='r2')

print("Resultados de cada fold:")
for i, score in enumerate(cv_scores):
    print(f'Fold {i+1}: R² = {score}')

# Imprimir el prom los 5 folds
print(f'R² promedio con validación cruzada (5 folds): {cv_scores.mean()}')
```
